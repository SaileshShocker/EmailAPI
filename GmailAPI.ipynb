```python
from __future__ import print_function
import base64  
import pandas as pd
import pickle
import os.path
from googleapiclient.discovery import build
from google_auth_oauthlib.flow import InstalledAppFlow
from google.auth.transport.requests import Request

# If modifying these scopes, delete the file token.pickle.
SCOPES = ['https://www.googleapis.com/auth/gmail.readonly']

def main():
    """Shows basic usage of the Gmail API.
    Lists the user's Gmail labels.
    """
    creds = None
    # The file token.pickle stores the user's access and refresh tokens, and is
    # created automatically when the authorization flow completes for the first
    # time.
    if os.path.exists('token.pickle'):
        with open('token.pickle', 'rb') as token:
            creds = pickle.load(token)
    # If there are no (valid) credentials available, let the user log in.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                'credentials.json', SCOPES)
            creds = flow.run_local_server(port=0)
        # Save the credentials for the next run
        with open('token.pickle', 'wb') as token:
            pickle.dump(creds, token)

    service = build('gmail', 'v1', credentials=creds)

    # Call the Gmail API
    # results = service.users().labels().list(userId='me').execute()
    # labels = results.get('labels', [])

    # get message
#     label = input('Which section do you want? ')
    label = 'INBOX'
    print('\n')
    results = service.users().messages().list(userId='me',labelIds=label).execute()
    messages = results.get('messages',[])
    
    
    # if not labels:
    message_count = int(input('How many emails do you want to extract? '))
    print('\n')
    if not messages:
#         print('No labels found.')
        print('No message found.')
    else:
#         print('Labels:')
#           print('Messages: ')
#         for label in labels:
          for message in messages[:message_count]:
            msg = service.users().messages().get(userId='me', id=message['id'], metadataHeaders=None).execute()
#             print(msg['payload']['body'].keys())
            

#             print(msg['payload']['headers'][1]['value'].split(';')[0].strip())

            print('From: ',msg['payload']['headers'][4]['value'].split(';')[0].strip())
            print('To: ',msg['payload']['headers'][5]['value'].split(';')[0].strip())
            print('DateTime: ',msg['payload']['headers'][1]['value'].split(';')[0].strip())
            print('Subject: ',msg['payload']['headers'][3]['value'].split(';')[0].strip())
                

        
            
            try:
                temp = base64.urlsafe_b64decode(msg['payload']['parts'][0]['body']['data'].encode('UTF8'))
                print('Body: ',temp.decode('UTF8'))
            except: 
#                 print(msg['payload']['parts'])
                print('Unable to fetch body!')
                
                          

            
            print('\n','__'*50)


if __name__ == '__main__':
    main()
```
